**Import Modules**

```python
import pandas as pd
import numpy as np
```

**Create a Dataframe**

```python
data1 = {'Name':pd.Series(['AAA','BBB','CCC','DDD','EEE'],index=['Reg000001','Reg000002','Reg000003','Reg000004','Reg000005']),
        'Mobile_No':pd.Series(['1111111','2222222','3333333','4444444','5555555'],index=['Reg000001','Reg000002','Reg000003','Reg000004','Reg000005']),
        'Marks':pd.Series([90,80,40,78,63],['Reg000001','Reg000002','Reg000003','Reg000004','Reg000005']),
        'Course':pd.Series(['OS','ML','BDA','NLP','MMAI'],index=['Reg000001','Reg000002','Reg000003','Reg000004','Reg000005']),
        'Status':pd.Series(['PASS','PASS','FAIL','PASS','PASS'],index=['Reg000001','Reg000002','Reg000003','Reg000004','Reg000005'])}

data2 = {'Name':pd.Series(['CCC','FFF'],index=['Reg000003','Reg000006']),
        'Mobile_No':pd.Series(['3333333'],index=['Reg000003']),
        'Marks':pd.Series([40,105],['Reg000003','Reg000006']),
        'Course':pd.Series(['OS','DA'],index=['Reg000003','Reg000006']),
        'Status':pd.Series(['FAIL','PASS'],index=['Reg000003','Reg000006'])}
df1 = pd.DataFrame(data1)
print(df1)
print('xxxxxxxxxxxxxxxxx')
df2 = pd.DataFrame(data2)
print(df2)
```

**Consolidated Dataframe**

```python
df = df1.append(df2)
df.index.name = 'Student ID'
print(df)
```

**Q1. Count the number of students in the frame**

```python
# noOfStudents = df.Name.describe()['unique']
noOfStudents = df.index.nunique()
print('Number of total entries of students: ' + str(len(df.index)))
print('Number of unique entries of students: ' + str(noOfStudents))
```

**Q2. Add a new column that should represent the grade letter (A>=90,B>=80,C>=70,D>=60,F<60) based on the marks obtained by students.**

```python
conditions = [
    (df['Marks'] < 60),
    (df['Marks'] >= 60) & (df['Marks'] < 70),
    (df['Marks'] >= 70) & (df['Marks'] < 80),
    (df['Marks'] >= 80) & (df['Marks'] < 90),
    (df['Marks'] >= 90)]
values = ['F','D','C','B','A']
df['Grade'] = np.select(conditions, values)
print(df)
```

**Q3. Insert the missing data with proper value input from user.**

```python
# print(df.isnull())
df = df.fillna(input("Enter mobile number for missing entry: "))
print(df)
```

**Q4. Remove the duplicate rows and replace the names with your friends names.**

Remove duplicate rows

```python
df = df[~df.index.duplicated(keep='first')]
print(df)
```

Replace Names with friends's name.

```python
df=df.replace('AAA','Ali')
df=df.replace('BBB','Babar')
df=df.replace('CCC','Cheema')
df=df.replace('DDD','Daud')
df=df.replace('EEE','Ejaz')
df=df.replace('FFF','Farhan')
print(df)
```

**Q5. Let us assume that teacher decided to upgrade the result of students that lack 2 marks from their result to promote them to next grade. Check and promote such students and print new result.**

```python
conditions = [
    (df['Marks'] < 58),
    (df['Marks'] >= 58) & (df['Marks'] < 68),
    (df['Marks'] >= 68) & (df['Marks'] < 78),
    (df['Marks'] >= 78) & (df['Marks'] < 88),
    (df['Marks'] >= 88)]
values = ['F','D','C','B','A']
df['Grade'] = np.select(conditions, values)
print(df)
```

**Q6. Mistakenly, the course name was entered wrongly by teacher, Replace the course “DA” with “BDA” and print updated frame to user.**

```python
df=df.replace('DA','BDA')
print(df)
```

**Q7. Display a list of students who achieved grade “B”.**

```python
students_b = df.loc[df['Grade'] == 'B']
# print(students_b['Name'].to_string(index=False))
print(students_b['Name'])
```

**Q8. Validate the marks for the range 0 to 100 so that any marks that were entered wrongly could be rectified.**

```python
conditions = [
    (df['Marks'] < 0),
    (df['Marks'] >= 0) & (df['Marks'] <= 100),
    (df['Marks'] > 100)
    ]
values = [0,df['Marks'],100]
df['Marks'] = np.select(conditions, values)
print(df)
```
